## Column name cleaning

```python
import pandas as pd
```

```python
general_data = pd.read_excel('LaligaGeneralStats24_25.xlsx')
defense_data = pd.read_excel('LaligaDefenseStats24_25.xlsx')
passing_data = pd.read_excel('LaligaPassingStats24_25.xlsx')
possesion_data = pd.read_excel('LaligaPossesionStats24_25.xlsx')
shooting_data = pd.read_excel('LaLigaShootingStats24_25.xlsx')
shotcreation_data = pd.read_excel('LaLigaShotCreationStats24_25.xlsx')
```

```python
def clean_columns(df):
    """Function to clean column up column names, removing spaces and special characters"""
    df.columns = (df.columns.str.lower()
                  .str.replace('+', '_plus_')
                  .str.replace('-', '_minus_')
                  .str.replace(' ', '_')
                  .str.replace('%', '_pct_')
                  .str.replace('/', '_per_')
                  .str.replace(':', '')
                 )
    df = df.rename(columns = {'90s' : 'ninety_s', 'min' : 'minutes'})
    return df
```

```python
general_data = clean_columns(general_data)
defense_data = clean_columns(defense_data)
passing_data = clean_columns(passing_data)
possesion_data = clean_columns(possesion_data)
shooting_data = clean_columns(shooting_data)
shotcreation_data = clean_columns(shotcreation_data)
```
